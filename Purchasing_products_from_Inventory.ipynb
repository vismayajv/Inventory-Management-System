```python
import json
```

```python
fd = open("data.json","r")
txt = fd.read()
fd.close()
data = json.loads(txt)

print("****************************************************************************************************")
print("*************************************INVENTORY MANAGEMENT SYSTEM************************************")
print("****************************************************************************************************\n")

for i in data.keys():
    print(i,data[i])
product_id = str(input("\n\nEnter the product id: "))
product_qty = int(input("Enter the Quantity: "))
if (data[product_id]['Product_Qty'] >= product_qty):  
  customer_name=str(input("\nEnter customer name : "))

  if (data[product_id]['Product_Qty'] >= product_qty):
    print("Name: ",data[product_id]['Product_Name'])
    print("Price: ",data[product_id]['Product_Price'])
    print("Quantity: ",product_qty)
    print("Manufacture_Date :",data[product_id]['Manufacture_Month and Year'])
    print("Expire_Date :",data[product_id]['Expire_Date'])
    print('\n-------------------------INVOICE----------------------------')
    print("Customer Name: ",customer_name)
    print("Billing Amount: ",data[product_id]['Product_Price'] * product_qty)
    print("Thanks for Purchasing \nVisit Us Again")
    print('--------------------------------------------------------------')
    
    data[product_id]['Product_Qty'] = data[product_id]['Product_Qty'] - product_qty  

                    
else:                                                                       
  print("Sorry, We are not having that much of quantity.")
  print("We're only having",data[product_id]['Product_Qty'],"quantity.")
  print("Would you like to purchase or not (Y/N)")
  ch = str(input("Y/N: "))
  if (ch == "Y"):     
    customer_name=str(input("\nEnter customer name : "))                                                     
    print("Name:",data[product_id]['Product_Name'])
    print("Price:",data[product_id]['Product_Price'])
    print("Quantity:", data[product_id]['Product_Qty'])
    print('\n-------------------------INVOICE---------------------------')
    print("Customer Name: ",customer_name)
    print("Billing Amount: ",data[product_id]['Product_Price'] * product_qty)
    print("Thanks for Purchasing \nVisit Us Again")
    print('--------------------------------------------------------------')

    data[product_id]['Product_Qty'] = 0 
        
        
  else:
    print("Thank You for Visiting Us")

          
```
